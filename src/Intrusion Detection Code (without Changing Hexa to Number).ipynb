```python
X, y = awid.select_dtypes(['number']), awid['class']
```

```python
# read in the test data and do the same transformations to it to match the training data

dtype=object 

##pd.set_option('display.max_rows', 500)  
##pd.set_option('display.max_columns', 500)

result = pd.read_csv('testdata.csv',low_memory=False)

##reading column headers
text_file = pd.read_csv('column names.csv',sep='\t', header=None, low_memory=False) ##reading file with column names
text_file.columns=['filename']
column_name_arr = text_file.filename.get_values() ##converting column names into an array 

##assigning column headers
result.columns= column_name_arr

#data processing
awid_test = result

# replace ? with None
awid_test.replace({"?": None}, inplace=True)

#drop the problematic columns
# goal is to remove columns that have over 10% of missing data
columns_with_mostly_null_test_data = awid_test.columns[awid_test.isnull().mean() >= 0.5]

#drop columns with more than 50% of missing data

awid_test.drop(columns_with_mostly_null_test_data, axis=1, inplace=True)
for col in awid_test.columns:
      awid_test[col] = pd.to_numeric(awid_test[col], errors='ignore')

```

```python
awid_test=awid_test.fillna(0) 
```

```python
# basic metric, accuracy

X_test = awid_test.select_dtypes(['number'])
y_test = awid_test['class']

```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier()
start = timeit.default_timer()
forest.fit(X, y)
y_preds = forest.predict(X_test)      
stop = timeit.default_timer()
print('Time: ', stop - start)  

print('Accuracy score '+ str(accuracy_score(y_test, y_preds)))
print ('Precision ' + str(precision_score(y_test, y_preds,average='weighted')))
print ('Recall score' + str(recall_score(y_test, y_preds,average='weighted')))
print ('f1 score' + str(f1_score(y_test, y_preds,average='weighted')))
print ('Confusion matrix:') 
print(confusion_matrix(y_test, y_preds))

        


```
